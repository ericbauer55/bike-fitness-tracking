```python
import pandas as pd
pd.set_option("display.max_columns", 100)
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from typing import Optional, Any, Union
```

```python
IN_CSV_DATA = Path().cwd().parent.parent / "data/3_feature_engineered"
OUT_CSV_DATA = Path().cwd().parent.parent / "data/4_data_split"
```

```python
df_summary = pd.read_csv(IN_CSV_DATA/'ride_summary_selected_feature.csv')
df_summary['start_time'] = pd.to_datetime(df_summary['start_time'])
# Sort the summary by ascending order of ride start time
df_summary = df_summary.set_index('start_time').sort_index(ascending=True).reset_index()
```

```python
df_summary.head()
```

```python
from sklearn.model_selection import train_test_split
df_train, df_test = train_test_split(df_summary, test_size=0.1, shuffle=False)
print(f'There are {df_train.shape[0]} total training examples ending at date: {df_train.loc[df_train.shape[0]-1, 'start_time']}')
print(f'There are {df_test.shape[0]} total training examples beginning at date: {df_test.reset_index().loc[0, 'start_time']}')
```

```python
df_train.to_csv(OUT_CSV_DATA / 'ride_summary_train.csv', index=False)
df_test.to_csv(OUT_CSV_DATA / 'ride_summary_test.csv', index=False)
```
